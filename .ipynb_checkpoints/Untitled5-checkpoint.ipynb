```python
import pandas as pd
from collections import Counter
orgdata = pd.read_csv("2022台灣就業通網站職缺清單2.csv")
orgdata.head()
```

```python
subdata = orgdata[['COMPNAME', 'EDGRDESC', 'seg']]
output_series = subdata.groupby(['EDGRDESC'])['seg'].apply(lambda x: ','.join(x)).reset_index()
output_series
```

```python
output_series['seg'] = output_series['seg'].str.replace("', '", " ")
output_series
```

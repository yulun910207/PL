```python
#str練習
date="2019/01/15"
topic="損害賠償"
print("日期",date,"\n判決內容",topic)
print("判決書內容:",'''
臺灣高等法院臺南分院民事判決　107年度上易字第163號　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　
上　訴　人　葉宏斌　
輔　佐　人　朱乾龍　
被上訴人　　黃榮義　

訴訟代理人　陳明律師　
　　　　　　陳偉仁律師
上列當事人間請求損害賠償事件，上訴人對於中華民國107年4月
27日臺灣嘉義地方法院第一審判決（107年度訴字第167號）提起
上訴，本院於107年12月25日言詞辯論終結，判決如下：
    主    文
上訴駁回。
第二審訴訟費用由上訴人負擔。
    事實及理由
一、上訴人主張：被上訴人於民國105年4月7日上午8時許，駕駛
    自用小貨車，沿嘉義縣太保市埤鄉里村里道路，由西往東方
    向行駛，途經同縣市○○○００號前時，本應注意車前狀況，
    並隨時採取必要之安全措施，且依當時天候、路況良好，並
    無不能注意之情形，竟疏未注意，冒然行駛，適伊駕駛自用
    小客車，沿同路由南往北方向行駛，二車發生擦撞(下稱系爭
    車禍)，致伊受有頸部肌肉扭傷及拉傷、頸部頸椎椎間盤移
    位等傷害，經復健治療無效。系爭車禍發生時，伊雖外表無
    明顯外傷，但於隔日睡醒後發覺脖子酸痛，而於105年4月18
    日到戴德森醫療財團法人嘉義基督教醫院(下稱嘉基醫院)就
    診，並確認伊因系爭車禍受頸椎甩鞭傷外力，引起頸椎脊髓
    損傷及椎間盤突出，且於同年5月13日經頸椎核磁共振證實
    ，可見伊受傷與被上訴人之過失行為間具有相當因果關係。
    爰依民法第184條第1項前段、第193條第1項、第195條第1項
    之規定，請求被上訴人給付新臺幣（下同）150萬元(包括醫
    藥費75,000元、減少勞動能力損失925,000元及精神慰撫金
    50萬元)。原審為伊敗訴之判決，尚有未洽。並聲明：(一)
    原判決廢棄。(二)被上訴人應給付伊150萬元及自起訴狀繕
    本送達之翌日起至清償日止，按週年利率5%計算之利息。(
    三)願供擔保，請准宣告假執行。
二、被上訴人則以：系爭車禍發生於105年4月7日上午8時許，上
    訴人於車禍當天警詢時已表示無受傷，嗣雖改稱車禍隔日發
    覺脖子痠痛，惟上訴人於疼痛發生11日後始至醫院就診，尚
    難認上訴人於105年5月16日開立診斷證明書之傷勢與系爭車
    禍有因果關係。又上訴人於系爭車禍後1年6個月之後，始發
    生診斷證明書所載「右側第五肋骨骨折、右側大腿挫傷，右
    肩膀、右手肘及右踝擦傷」之傷害，均非105年4月7日傷勢
    之延伸，故上訴人107年1月31日傷害結果與系爭車禍並無因
    果關係，若系爭車禍發生後上訴人果真發生傷害程度十分嚴
    重之頸椎椎間盤移位，殊無可能遲至106年10月9日始手術治
    療等語，資為抗辯。並聲明：上訴駁回。
三、兩造不爭執事項：    
    被上訴人於105年4月7日上午8時許，駕駛自用小貨車，沿嘉
    義縣太保市埤鄉里村里道路，由西往東方向行駛，途經同縣
    市○○○００號前時，與上訴人駕駛自用小客車發生車禍。
四、上訴人主張被上訴人有侵權行為，致其受有損害，爰請求損
    害賠償，惟為被上訴人所否認，並以上情置辯；是上訴人之
    損害與系爭車禍間有無因果關係？上訴人請求被上訴人給付
    150萬元本息，有無理由？厥為本件應審究之爭點，爰分述
    如下：
(一)上訴人之損害與系爭車禍間無因果關係：
 1、上訴人固主張其受有頸部肌肉扭傷及拉傷、頸部頸椎椎間盤
    移位等傷害，經復健治療無效果，於106年10月7日急診住院
    ，同年月9日手術後，大小便失禁，已難痊癒云云，並提出
    嘉基醫院所出具之診斷證明書及收據為證（原審卷第21至23
    頁）。然觀諸該收據之記載，上訴人係於106年10月7日住院
    ，同年月9日開刀而支出醫療費用，住院之日期為106年10月
    7日至17日，然此次之醫療支出，距105年4月7日發生系爭車
    禍之日期，已有1年6個月，上訴人之主張，已非無疑。而依
    上訴人之自承，其於106年10月7日另與他人發生車禍(本院
    卷第71頁，下稱第二次車禍)，並經本院調閱第二次車禍之
    警卷(本院卷第97至109頁)核閱屬實，上訴人並與第二次車
    禍之對造達成和解，由對造賠付其含醫療費等費用共90萬元
    ，亦有調解書可參(本院卷第155頁)。故第二次車禍發生於1
    06年10月7日，上訴人提出之前揭求償依據之診斷證明書，
    依醫師囑言之記載「病患（指上訴人，下同）因上述病於10
    6年10月7日13時37分到急診」，顯見上訴人本件請求所提出
    之診斷證明書及醫療費收據，係因上訴人第二次車禍受傷就
    醫而產生之費用，上訴人據以向被上訴人請求損害賠償，已
    嫌無據。
２、上訴人雖另提出其受有頸部扭拉傷、頸部頸椎椎間盤移位傷
    害之診斷證明書及MRI影像資料，其中醫師囑言記載：「病
    患因上述病因於105年4月18日、5月3日、5月16日、6月1日
    、6月22日、7月22日至神經外科門診就診。建議手術(開刀)
    治療」，並主張依MRI影像資料，其於105年5月13日核磁共
    振檢查有第5、6節頸椎椎間盤突出等情(原審卷第17、75頁
    、本院卷第83、85、89頁)。惟系爭車禍於105年4月7日發生
    ，上訴人遲至同年月18日始就診，距車禍發生時已相隔11日
    ，同年5月13日核磁共振檢查，第5、6節頸椎椎間盤突出之
    傷害，距車禍發生已相隔一個多月，上訴人前述傷害是否因
    系爭車禍所造成，已非無疑。另參諸上訴人於105年4月7日
    車禍當天於警詢中陳稱：伊及對方都沒有受傷等語；嗣於同
    年5月4日警詢中改稱：伊於發生交通事故當下向警方表示伊
    及對方沒有受傷，隔日睡醒後才發覺脖子痠痛，伊於105年4
    月18日有至嘉基醫院就診等語；復於同年5月24日偵查陳稱
    ：伊車禍當時有告訴警察說伊沒有受傷，車禍當天伊沒有去
    就醫，後來伊覺得脖子很痛，105年4月18日才去就醫等語(
    見臺灣嘉義地方檢察署105年度偵字第3394號偵查卷及所附
    警卷)，故上訴人原陳稱並未受傷，則其於105年4月18日或
    同年5月13日所受之傷害，究竟是否因系爭車禍所造成，亦
    非無疑。
３、上訴人於本院雖另提出嘉基醫院診斷證明書，依醫師囑言之
    記載「病人因105年4月7日車禍時因頸椎甩鞭傷外力，引起
    頸椎脊髓損傷及椎間盤突出，已於105年5月13日頸椎核磁共
    振證實，並於106年10月9日施行第5到7節頸椎椎間盤突出切
    除手術，目前在本院持續接受術後治療及復健治療中」(原
    審卷第213頁、本院卷第17、81頁)，惟本件診斷證明書顯與
    事實不符，如前所述，上訴人於106年10月7日另與他人發生
    第二次車禍，其於同年10月7日急診住院，同年月9日手術，
    至同年月17日出院，該等急診住院手術，均係第二次車禍所
    造成，惟前揭診斷證明書卻記載該等住院手術係來自105年4
    月7日系爭車禍時因頸椎甩鞭傷外力，所引起之頸椎脊髓損
    傷及椎間盤突出而手術，診斷證明書此項記載已屬有誤。且
    醫師究竟如何判斷上訴人於106年10月9日手術所治療之傷害
    ，其傷係來自105年4月7日系爭車禍所造成，亦非無疑，此
    項因果關係之判斷，顯逾醫師所得判斷之範圍，經本院函詢
    嘉基醫院究竟如何判斷上訴人於106年10月9日之手術係因10
    5年4月7日系爭車禍受傷所造成，依該院之回函，指出上訴
    人於105年4月18日檢查，有第5、6節椎間盤退化；同年5月
    13日核磁共振檢查有第5、6節頸椎椎間盤突出，至於車禍因
    果關係部分則稱「甩鞭傷只能從病史推論，MRI(核磁共振)
    符合診斷」等情(本院卷第167頁)，惟依該函所述仍無明確
    之判斷依據，究竟如何從病史推論，亦失之空泛，且系爭車
    禍與106年10月9日之手術相隔1年半，其間上訴人亦曾發生
    第二次車禍，甚至造成急診住院手術之結果，亦難單憑病史
    推論，而不考慮時間及第二次車禍之事實因素，自難僅憑前
    揭函即認定上訴人於106年10月9日手術時之傷害係因系爭車
    禍所造成。
４、綜上各情以觀，上訴人於106年10月9日手術時之傷害，與系
    爭車禍已相隔1年半，且中間發生第二次車禍，該等傷害應
    係第二次車禍所造成，並非系爭車禍所造成；而上訴人雖於
    105年4月18日檢查，有第5、6節椎間盤退化；同年5月13日
    核磁共振檢查有第5、6節頸椎椎間盤突出等情，惟椎間盤退
    化顯與系爭車禍無關，而上訴人確診第5、6節頸椎椎間盤突
    出之時間係105年5月13日，距系爭車禍發生之時已一個多月
    ，亦難遽認上訴人所受之傷害與系爭車禍有關。
(二)上訴人請求被上訴人給付150萬元本息，並無理由：
    按故意或過失，不法侵害他人之權利者，負損害賠償責任；
    不法侵害他人之身體或健康者，對於被害人因此喪失或減少
    勞動能力或增加生活上之需要時，應負損害賠償責任；不法
    侵害他人之身體、健康，被害人雖非財產上之損害，亦得請
    求賠償相當之金額。民法第184條第1項前段、第193條第1項
    、第195條第1項前段分別定有明文，故侵權行為之成立，須
    行為人因故意或過失不法侵害他人權利，且不法行為與損害
    間有因果關係，始能成立。本件上訴人所受之傷害並無法證
    明係系爭車禍所造成，上訴人請求被上訴人給付醫藥費75,0
    00元、減少勞動能力損失925,000元及精神慰撫金50萬元，
    合計150萬元，自屬無據。上訴人主張被上訴人因本件車禍
    應負完全過失責任，聲請另送鑑定云云，自無必要，併予敘
    明。
五、綜上所述，上訴人依民法第184條第1項前段、第193條第1項
    、第195條第1項之規定，請求被上訴人給付150萬元及自起
    訴狀繕本送達之翌日起至清償日止，按週年利率5%計算之利
    息，為無理由，應予駁回。原審為上訴人敗訴之判決，並駁
    回其假執行之聲請，核無不合。上訴意旨指摘原判決不當，
    求予廢棄改判，為無理由，應駁回其上訴。
六、本件事證已臻明確，兩造其餘攻擊、防禦方法及舉證，核與
    判決之結果不生影響，爰不逐一論述，併此敘明。
七、據上論結，本件上訴為無理由，依民事訴訟法第449條第1項
    、第78條，判決如主文。
中    華    民    國  108    年    1    月    15  日
                   民事第四庭    審判長法  官  翁金緞
 
                                       法  官  張世展
 
                                       法  官  黃義成
 
上為正本係照原本作成。
不得上訴。 
中    華    民    國  108    年    1    月    15  日
 
                                      書記官  施淑華''')
```
